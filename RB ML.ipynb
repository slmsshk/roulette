```python
cheating = pd.read_excel( 'Roulet_cheating_pattern_set (1).xlsx')
rows,cols = cheating.shape
print(f'Total\nRows: {rows}\ncols: {cols}')
cheating['Fraud/Non_Fraud'] = 1
cheating.drop('WIN_NO',axis=1,inplace=True)
cheating.head()
```

```python
non_cheating = pd.read_excel( 'roulet_non_cheating_dataset.xlsx')
rows,cols = non_cheating.shape
print(f'Total\nRows: {rows}\ncols: {cols}')
non_cheating['Fraud/Non_Fraud'] = 0
non_cheating.drop(['TRANS'],axis=1,inplace=True)
first_3_columns = list(non_cheating.columns[:5]) 
update_cols = [f'NOS{i}' for i in non_cheating.columns[5:-1] ]
first_3_columns.extend(update_cols)
first_3_columns.append(non_cheating.columns[-1])
non_cheating.columns = first_3_columns
non_cheating.rename({'NOSO':'NOS0','NOSOO':'NOS00','Draw ID':'DRAW_ID','MEMBER ID':'MEMBER_ID','Date':'TRANS_DATE','SUM':'TOTAL_BET_AMOUNT'},axis=1,inplace=True)
non_cheating.head()
```

```python
compiled = pd.concat([cheating,non_cheating],axis=0).reset_index(drop=True)
compiled.to_csv('compiled_tuple.csv',index=False)
compiled[['CCS_ID','DRAW_ID']] = compiled[['CCS_ID','DRAW_ID']].astype('object')
compiled.head()
```

```python
agg_data = compiled.groupby('DRAW_ID').agg({
    'MEMBER_ID': lambda x: list(set(x)),
    'CCS_ID': lambda x: list(set(x))
}).reset_index()
agg_data
compiled['same_draw_id_diff_mem_id_n__casionid'] = 0

for _, row in agg_data.iterrows():
    condition = len(row['MEMBER_ID']) > 1 and len(row['CCS_ID']) > 1
    # Update the flag in the original DataFrame based on the condition
    compiled.loc[compiled['DRAW_ID'] == row['DRAW_ID'], 'same_draw_id_diff_mem_id_n__casionid'] = int(condition)

# Verify the operation
compiled[['DRAW_ID', 'same_draw_id_diff_mem_id_n__casionid']].head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y)
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(enable_categorical=True)
xgb.fit(x_train,y_train)
pred = xgb.predict(x_test)
xgb.score(x_test,y_test)*100
```

```python
import pickle

with open('test_model.pkl','wb') as f:
    pickle.dump(xgb,f)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import matplotlib.pyplot as plt
import numpy as np

matrix = confusion_matrix(y_test, pred)

# Initialize ConfusionMatrixDisplay with the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=matrix)

# Customize the plot
fig, ax = plt.subplots(figsize=(8, 8))  # Set figure size for better readability
disp.plot(cmap='Blues', ax=ax)  # Choose a color map

# Add titles and labels for clarity
ax.set_title('Confusion Matrix with Annotations', fontsize=16)
ax.set_xlabel('Predicted Labels', fontsize=14)
ax.set_ylabel('True Labels', fontsize=14)

# Annotations for TP, TN, FP, FN
labels = ['TN', 'FP', 'FN', 'TP']
counts = ["{0:0.0f}".format(value) for value in matrix.flatten()]
percentages = ["{0:.2%}".format(value) for value in matrix.flatten()/np.sum(matrix)]
annotations = [f"{label}\n\n{percentage}" for label, count, percentage in zip(labels, counts, percentages)]

# Position annotations in the center of each quadrant
threshold = matrix.max() / 2.
for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
        ax.text(j, i, annotations[i * 2 + j],
                 ha="center", va="center",
                 color="white" if matrix[i, j] > threshold else "black", fontsize=12)

# Optional: Customize tick labels size
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Optional: Add grid for easier interpretation
ax.grid(False)  # Adjust based on preference

# Save the plot to a file
plt.savefig('ConfusionMatrix_WithAnnotations.png', bbox_inches='tight')  # Save as PNG and ensure the layout is tight

# Show the plot
plt.show()
```
